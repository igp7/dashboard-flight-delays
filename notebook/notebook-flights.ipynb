```python
import random

n = 5819080 # Numero de filas del fichero
s = 10000 # Numero de filas del sample
filename = '../data/flight/flights.csv'
skip = sorted(random.sample(range(1, n-1),n-s))
flights_sample = pd.read_csv(filename, skiprows=skip)
flights_sample.head()
```

```python
airports = pd.read_csv('../data/flight/airports.csv', low_memory=False)
airports.head()
```

```python
airlines = pd.read_csv('../data/flight/airlines.csv', low_memory=False)
airlines.head()
```

```python
# flights_sample = flights.sample(n = 10_000)
flights_sample.drop(['CANCELLATION_REASON', 'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY'], axis=1, inplace=True)
flights_sample.shape
```

```python
flights_airports_origin = flights_sample.merge(airports, how='inner', left_on='ORIGIN_AIRPORT', right_on='IATA_CODE')
flights_airports_origin.rename({'AIRPORT':'ORIGIN_AIRPORT_NAME', 'CITY':'ORIGIN_CITY', 'STATE':'ORIGIN_STATE',
                        'COUNTRY': 'ORIGIN_COUNTRY', 'LATITUDE':'ORIGIN_LATITUDE', 'LONGITUDE':'ORIGIN_LONGITUDE'}, axis=1, inplace=True)
flights_airports_origin.head()
```

```python
flights_airports = flights_airports_origin.merge(airports, how='inner', left_on='DESTINATION_AIRPORT', right_on='IATA_CODE')
flights_airports.rename({'AIRPORT':'DESTINATION_AIRPORT_NAME', 'CITY':'DESTINATION_CITY', 'STATE':'DESTINATION_STATE',
                        'COUNTRY': 'DESTINATION_COUNTRY', 'LATITUDE':'DESTINATION_LATITUDE', 'LONGITUDE':'DESTINATION_LONGITUDE'}, axis=1, inplace=True)
flights_airports.head()
```

```python
flights_airports.drop(['IATA_CODE_x', 'IATA_CODE_y'], axis=1, inplace=True)
```

```python
flights_airlines = flights_airports.merge(airlines, how='inner', left_on='AIRLINE', right_on='IATA_CODE', suffixes = ('_CODE', '_NAME'))
flights_airlines.head()
```

```python
flights_airlines.dropna(inplace=True)
```

```python
flights_airlines.to_csv('../data/flight/vuelos.csv')
```
